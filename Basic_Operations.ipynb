```python
# Drawing a Line 
import numpy as np
blank_screen = np.zeros((360,560,3))
random_color = (250,250,250)
#cv2_imshow(blank_screen)
cv.line(blank_screen,(0,0),(280,180),random_color)
cv.line(blank_screen,(280,180),(560,360),random_color)
cv.line(blank_screen,(560,0),(280,180),random_color)
cv.line(blank_screen,(280,180),(0,360),random_color)
cv2_imshow(blank_screen)
print ("\n")

# Drawing a Rectangle
cv.rectangle(blank_screen,(100,100),(180,180),(250,0,0),-1)
cv.rectangle(blank_screen,(290,100),(90,180),(0,0,250))
cv2_imshow(blank_screen)
print ("\n")

# Drawing some concentric circles 
(xC,yC) = (int(blank_screen.shape[1]/2),int(blank_screen.shape[0]/2))
for i in range(0,200,25):
  cv.circle(blank_screen,(xC,yC),i,(0,250,0))
cv2_imshow(blank_screen)
```
